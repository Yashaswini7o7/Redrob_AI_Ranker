```python
import json
import re
import pandas as pd

from tqdm import tqdm
```

```python
candidates=[]

with open("/content/drive/MyDrive/Redrob-Hackathon/data/candidates.jsonl","r",encoding="utf-8") as f:

    for line in tqdm(f):

        candidates.append(json.loads(line))
```

```python
RETRIEVAL_KEYWORDS = {

"retrieval",

"retriever",

"hybrid search",

"dense retrieval",

"sparse retrieval",

"semantic search",

"document retrieval",

"information retrieval",

"query expansion"

}
```

```python
RANKING_KEYWORDS = {

"ranking",

"ranker",

"recommendation",

"recommendation engine",

"candidate matching",

"matching engine",

"learning to rank",

"ltr",

"ndcg",

"map",

"mrr"

}
```

```python
EMBEDDING_KEYWORDS = {

"embedding",

"embeddings",

"sentence transformer",

"sentence-transformers",

"e5",

"bge",

"openai embedding",

"vector embedding"

}
```

```python
VECTOR_DB = {

"faiss",

"pinecone",

"milvus",

"qdrant",

"weaviate",

"elasticsearch",

"opensearch"

}
```

```python
LLM = {

"llm",

"gpt",

"bert",

"transformer",

"fine tuning",

"finetuning",

"lora",

"qlora",

"peft"

}
```

```python
PYTHON = {

"python",

"pytorch",

"tensorflow",

"keras",

"numpy",

"pandas"

}
```

```python
def keyword_count(text, keyword_set):

    count = 0

    text = text.lower()

    for word in keyword_set:

        if word in text:

            count += 1

    return count
```

```python
def career_text(candidate):

    text = []

    text.append(candidate["profile"]["summary"])

    text.append(candidate["profile"]["headline"])

    for job in candidate["career_history"]:

        text.append(job["title"])

        text.append(job["description"])

    return " ".join(text).lower()
```

```python
rows = []

for c in tqdm(candidates):

    text = career_text(c)

    row = {}

    row["candidate_id"] = c["candidate_id"]

    row["retrieval_score"] = keyword_count(
        text,
        RETRIEVAL_KEYWORDS
    )

    row["ranking_score"] = keyword_count(
        text,
        RANKING_KEYWORDS
    )

    row["embedding_score"] = keyword_count(
        text,
        EMBEDDING_KEYWORDS
    )

    row["vector_db_score"] = keyword_count(
        text,
        VECTOR_DB
    )

    row["llm_score"] = keyword_count(
        text,
        LLM
    )

    row["python_score"] = keyword_count(
        text,
        PYTHON
    )

    rows.append(row)
```

```python
career_df = pd.DataFrame(rows)

career_df.head()
```

```python
career_df.to_csv(
"/content/drive/MyDrive/Redrob-Hackathon/output/career_features.csv",
index=False
)
```
