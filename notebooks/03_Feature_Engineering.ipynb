```python
def experience_score(years):

    if 5<=years<=8:

        return 100

    elif 4<=years<5:

        return 90

    elif 8<years<=10:

        return 85

    elif 3<=years<4:

        return 70

    else:

        return 40
```

```python
def title_score(title):

    title=title.lower()

    if "ai engineer" in title:

        return 100

    elif "ml engineer" in title:

        return 95

    elif "machine learning" in title:

        return 95

    elif "data scientist" in title:

        return 90

    elif "backend engineer" in title:

        return 60

    elif "software engineer" in title:

        return 70

    else:

        return 20
```

```python
def company_score(company):

    if company in PRODUCT_COMPANIES:

        return 100

    elif company in SERVICE_COMPANIES:

        return 40

    else:

        return 60
```

```python
def skill_score(skills):

    score=0

    for s in skills:

        skill=s["name"].lower()

        if skill in AI_SKILLS:

            score+=5

    return min(score,100)
```

```python
def get_career_text(candidate):

    text=[]

    text.append(candidate["profile"]["summary"])

    text.append(candidate["profile"]["headline"])

    for job in candidate["career_history"]:

        text.append(job["description"])

    return " ".join(text).lower()
```

```python
rows=[]
```

```python
for c in tqdm(candidates):

    profile=c["profile"]

    signals=c["redrob_signals"]

    row={}

    row["candidate_id"]=c["candidate_id"]

    row["experience_score"]=experience_score(
        profile["years_of_experience"]
    )

    row["title_score"]=title_score(
        profile["current_title"]
    )

    row["company_score"]=company_score(
        profile["current_company"]
    )

    row["skill_score"]=skill_score(
        c["skills"]
    )

    row["career_text"]=get_career_text(c)

    rows.append(row)
```

```python
feature_df=pd.DataFrame(rows)

feature_df.head()
```
